# **Часть 1**
## **Задание**: 
"В рамках сессионного задания необходимо развернуто ответить на вопрос: «Какие ключевые шаги необходимо выполнить для планирования и проведения MVT-теста?». В рамках этого задания требуется описать этапы подготовки, запуска и анализа многовариантного тестирования."

## **Ответ**

Согласно материалам курса, MVT (Multivariate Testing) - тестирование нескольких факторов (A, B, C) с несколькими вариациями (A1, A2, B1, B2...) одновременно. Целью MVT-тестирования является определение не только лучших комбинаций, но и взаимодействия факторов.

#### **Основные этапы MVT-тестирования**:

#### 1. Этап подготовки и планирования
1.1. Определение цели и ключевой метрики (KPI)

Цель: Должна быть конкретной, измеримой и согласованной с бизнес-стратегией. Например, для подразделения Банка, занимающегося активацией и удержанием клиентов сегмента "Малый и микробизнес": «Определить, какие комбинации изменений в процессе онбординга и раннего обслуживания клиентов ММБ наиболее эффективно снижают долю клиентов в оттоке на 4-й месяц жизни" 

Ключевая метрика: Выбирается одна первичная метрика для оценки успеха теста. Это должна быть метрика, непосредственно связанная с целью (например, конверсия, средний доход на пользователя, кликабельность). Рекомендуется использовать бинарные или непрерывные метрики, поддающиеся стандартным статистическим тестам.
Для нашего примера ключевой метрикой будет доля клиентов, находящихся в оттоке, на 4-й месяц жизни (ранним оттоком в подразделении Банка на основании наблюдений (эмпирически) принято считать долю клиентов в оттоке на 4-й месяц жизни).

Клиент считается находящимся в оттоке, если в течение 3-х месяцев, включая отчетный, у него выполняются все условия ниже:
- не было операций по расчетному счету
- нет остатков по депозиту (или остатки менее 1 тыс. руб.)
- нет действующего кредита (или есть просрочки 90+).
    Также уточним, что сравнивать будем однородные когорты (т.е. клиентов, привлеченных в один и тот же месяц) для исключения сезонности и макроэкономических факторов.

    Также будем смотреть на вторичные (контрольные) метрики за первые 4 месяца жизни:
    - среднее кол-во транзакций на клиента
    - средняя сумма оборотов на клиента 
    - средний фактический доход на клиента 
    - среднее кол-во негативных триггеров на клиента и их структура
    Под негативными триггерами подразумевается причина недовольства клиента работой с Банком, выявленная либо в голосовых коммуникациях, либо в чате.

1.2. Выбор процесса и факторов (элементов) для теста

Процесс для теста: на примере с клиентами ММБ - процесс первых 90 дней обслуживания клиента ММБ после открытия счета. Это ключевой период для активации и предотвращения раннего оттока.

Факторы (элементы вмешательства): определяются независимые переменные — конкретные изменения в процессе взаимодействия с клиентом. Каждому фактору назначаются два или более уровня.

Пример для рассматриваемого банковского MVT:

Фактор A: Тип персонального менеджера:
A1: Выделенный персональный менеджер (офлайн).
A2: Удаленный менеджер (онлайн).
A3: Без выделенного менеджера (обслуживание через контакт-центр).

Фактор B: Мероприятия для ранней активации.

B1: Начисление бонуса в размере 1 000 руб. за раннюю активацию (например, за перевод на сумму 1 000 руб. и более).
B2: Возможность участия в розыгрыше призов за выполнение целевого действия (перевод/открытие депозита на сумму более 1 тыс. руб.)
B3: Пакет бесплатных переводов на 3 месяца

Фактор C: Частота коммуникаций в первый месяц.
C1: Высокая (звонок для целей онбординга + 2 персонализированных смс/email в неделю).
C2: Умеренная (приветственный звонок + 1 персонализированное смс/email в неделю).
C3: Низкая (только 1 персонализированное смс/email).

Обратим внимание, что на нашем примере для факторов с уровнями 3×3×3 полный факторный план составит 27 комбинаций. В подобных случаях, учитывая ограниченный поток клиентов, на этапе планирования может принимабыть принято решение об использовании дробного факторного плана *Fractional Factorial Design*, сокращающего число тестируемых комбинаций.

Например, для нашего случая, это может быть план L9(3^4). Это специальная ортогональная таблица (Orthogonal Array):
L9 — означает, что таблица содержит 9 строк (т.е. 9 уникальных тестовых комбинаций).
(3^4) — означает, что таблица позволяет тестировать до 4-х факторов, каждый из которых имеет ровно 3 уровня.

Как использовать эту таблицу: сопоставляем каждый из трёх факторов (например, в нашем случае менеджер, активация, коммуникации) с первыми тремя столбцами таблицы, а цифры 1, 2, 3 в ячейках трактуем как номера уровней для каждого фактора. Каждая строка полученной таблицы задаёт одну из 9 тестовых комбинаций, которые необходимо реализовать и на которые нужно случайным образом распределять новых клиентов.
Ортогональность таблицы гарантирует, что каждый уровень каждого фактора встречается в комбинации с каждым уровнем любого другого фактора ровно одинаковое количество раз.
Fractional Factorial Design позволяет за разумное время и стоимость получить ответ на вопрос: «Как в среднем влияет на отток сам по себе тип менеджера, тип активации и частота коммуникаций?». Т.е. дробный факторный анализ не позволяет определить наилучшие комбинации факторов. Для определения наилучших комбинаций можно сделать анализ в 2 этапа:
1. Определить наиболее влияющие факторы (с помощью плана L9(3^4))
2. Провести  полный или более насыщенный дробный план (например, 2×2×2) уже для меньшего числа факторов, чтобы точно оценить их взаимодействия и найти оптимальную комбинацию.


1.3. Формулировка гипотез 

Для MVT гипотезы формулируются как для главных эффектов (влияние каждого отдельного фактора), так и для эффектов взаимодействия (синергетическое или антагонистическое влияние комбинации факторов). Нулевая гипотеза (H₀) всегда утверждает об отсутствии эффекта, альтернативная (H₁) — о его наличии.

Для нашего MVT с тремя факторами (A, B, C):

A. Главные эффекты (для полного или дробного плана):

Фактор A (Тип менеджера):

H₀(A): Тип персонального менеджера (выделенный/удалённый/отсутствие) не оказывает статистически значимого влияния на долю оттока клиентов на 4-й месяц.

H₁(A): Тип персонального менеджера оказывает статистически значимое влияние на долю оттока. (Мы ожидаем, что выделенный менеджер (A1) приведёт к наименьшему оттоку благодаря персональному доверительному контакту, а его отсутствие (A3) — к наибольшему)

Фактор B (Мероприятие активации):

H₀(B): Тип мотивационного мероприятия для ранней активации не влияет на долю оттока.

H₁(B): Тип мероприятия статистически значимо влияет на отток. (Мы ожидаем, что денежный бонус (B1) будет наиболее эффективен для немедленного формирования привычки использования счета, снижая отток сильнее, чем стимулы B2 и B3)

Фактор C (Частота коммуникаций):

H₀(C): Частота персональных коммуникаций в первый месяц не влияет на долю оттока.

H₁(C): Частота коммуникаций статистически значимо влияет на отток. (Мы ожидаем оптимальный (умеренный) уровень контактов (C2), который будет эффективнее как недостаточного внимания (C3), так и чрезмерного, которое может восприниматься как спам и увеличивать отток (C1)).

B. Взаимодействия (актуальны для полного или уточняющего плана на 2-м этапе):

Взаимодействие A×B (Менеджер × Мероприятие):

H₀(A×B): Эффект от типа менеджера не зависит от типа мероприятия активации (взаимодействия нет).

H₁(A×B): Существует значимое взаимодействие. (Мы предполагаем, что денежный бонус (B1) будет особенно эффективен в паре с удалённым менеджером (A2), так как не требует сложных объяснений, а розыгрыш призов (B2) лучше сработает с выделенным менеджером (A1), который может лично вовлечь клиента)

Взаимодействие A×C (Менеджер × Коммуникации):

H₀(A×C): Влияние типа менеджера не зависит от частоты коммуникаций.

H₁(A×C): Влияние типа менеджера статистически значимо зависит от частоты коммуникаций. (Мы ожидаем, что для клиентов без выделенного менеджера (A3) более частые автоматические коммуникации (C1) могут компенсировать отсутствие персонального контакта и снизить отток)

Взаимодействие B×C:

H₀(B×C): Эффект мероприятия не зависит от частоты коммуникаций.

H₁(B×C): Существует статистически значимое взаимодействие. (Ожидаем, что B1 (бонус) эффективен даже при C3 (низкая частота), а B2/B3 требуют более частых контактов (C1/C2)).

Важное замечание: При использовании дробного плана L9(3^4) мы формально не можем проверить гипотезы о взаимодействиях, так как их эффекты «смешаны». Поэтому на первом (скрининговом) этапе мы фокусируемся на проверке H₁(A), H₁(B), H₁(C). Обнаруженные значимые главные эффекты станут основой для формулировки гипотез о взаимодействиях на втором, более детальном этапе тестирования.

1.4. Расчет размера выборки и длительности теста

Это критически важный шаг для обеспечения статистической мощности теста — вероятности обнаружить эффект, если он действительно существует. Недостаточный размер выборки ведёт к ложноотрицательным результатам (ошибка II рода), когда мы не увидим реального снижения оттока.

На примере нашего кейса:

Базовый уровень метрики: Историческая доля оттока на 4-й месяц = 25%.

Минимальный обнаруживаемый эффект (MDE): допустим, мы хотим обнаружить снижение оттока минимум на 3 процентных пункта. Это изменение имеет практическую значимость.

Уровень значимости (α): 0.05 (стандартный 5% риск ложноположительного результата).

Статистическая мощность (1-β): 0.8 (стандартные 80%).

Число групп (вариантов):

Для полного факторного плана: 27 комбинаций (3×3×3).
Для дробного плана L9 (Fractional Factorial): 9 комбинаций.

Расчет размера выборки (Python с библиотекой statsmodels):

Поскольку наша ключевая метрика — доля (биномиальное распределение), используем подход, основанный на тесте хи-квадрат для сравнения нескольких пропорций.
Приведем пример расчетов на python:

```python
import math
from statsmodels.stats.proportion import samplesize_proportions_2indep_onetail

# Параметры
p0 = 0.25          # Базовый уровень оттока
mde_abs = 0.03     # MDE = 3%
p1 = p0 - mde_abs  # целевой уровень оттока = 22%
alpha = 0.05      
power = 0.8        
k_full = 27        # 27 групп (полный план)
k_frac = 9         # 9 групп (дробный план L9)
monthly_clients = 45000  # среднее кол-во новых клиентов ММБ в месяц

# 1. Расчет размера выборки с учетом поправки Бонферрони
# для 27 групп:
alpha_bonferroni_full = alpha / k_full
n_bonferroni_full = samplesize_proportions_2indep_onetail(
    diff=mde_abs,
    prop2=p1,
    alpha=alpha_bonferroni_full/2,
    power=power,
    ratio=1
)
n_bonferroni_full = math.ceil(n_bonferroni_full)
# для 9 групп:
alpha_bonferroni_frac = alpha / k_frac
n_bonferroni_frac = samplesize_proportions_2indep_onetail(
    diff=mde_abs,
    prop2=p1,
    alpha=alpha_bonferroni_frac/2,
    power=power,
    ratio=1
)
n_bonferroni_frac = math.ceil(n_bonferroni_frac)

# 2. Расчет длительности теста (учитываем ожидание в 4 месяца с момента привлечения)

observation_months = 4  # Оцениваем отток на 4-й месяц жизни клиента

# Расчет времени набора клиентов
time_to_recruit_full = (n_bonferroni_full * k_full) / monthly_clients
time_to_recruit_frac = (n_bonferroni_frac * k_frac) / monthly_clients

# Общая длительность теста  = время набора + время наблюдения
total_duration_full = time_to_recruit_full + observation_months
total_duration_frac = time_to_recruit_frac + observation_months

print(f"Для полного плана (27 групп):")
print(f"• Нужно клиентов: {n_bonferroni_full * k_full}")
print(f"• Время набора: {time_to_recruit_full:.1f} месяцев")
print(f"• Время наблюдения: {observation_months} месяцев")
print(f"• Общая длительность теста: {total_duration_full:.1f} месяцев ({math.ceil(total_duration_full)} месяцев)")

print(f"\nДля плана L9 (9 групп):")
print(f"• Нужно клиентов: {n_bonferroni_frac * k_frac}")
print(f"• Время набора: {time_to_recruit_frac:.1f} месяцев")
print(f"• Время наблюдения: {observation_months} месяцев")
print(f"• Общая длительность теста: {total_duration_frac:.1f} месяцев ({math.ceil(total_duration_frac)} месяцев)")
```

Основываясь на результате расчета выше, для нашего примера можно рекомендовать принять к реализации план L9, т.к: 
- длительность теста менше на 3 месяца, следовательно, бизнес сможет быстрее принять меры по сокращению оттока
- требуется меньше ресурсов для проведения теста: 53 тыс. клиентов vs 186 тыс.
- ограниченное кол-во одновременных изменений снижает операционные риски

1.5. Техническая подготовка и стратегия рандомизации

Это этап создания инфраструктуры для корректного проведения эксперимента и определения метода распределения клиентов по тестовым группам. 
Для нашего кейса это означает настройку банковских систем (CRM, процессинг коммуникаций) для реализации плана L9 и выбор стратегии рандомизации (например, простая случайная или стратифицированная).

Для нашего кейса допустимо применение простой случайной рандомизации на основе детерминированного хеша от client_id. Это обусловлено достаточным объемом выборки (~53 тыс. клиентов), гарантирующим баланс групп по закону больших чисел, а также операционной сложностью стратификации для новых клиентов ММБ.

Также необходимо обеспечить сбор данных по всем метрикам (отток, транзакции, доход) и последующую проверку баланса групп по ключевым характеристикам (регион, канал привлечения) для подтверждения корректности рандомизации.

#### 2. Этап проведения теста

2.1. Запуск и мониторинг
Это этап непосредственной реализации запланированных изменений для выбранной аудитории, сбор первичных данных и постоянный контроль за корректностью работы системы.

В нашем кейсе запуск означает включение алгоритма распределения для всех новых клиентов ММБ, которые начинают попадать в одну из 9 предопределённых групп согласно плану L9. Система CRM/маркетинга автоматически назначает клиенту тип менеджера (A), мероприятие активации (B) и частоту коммуникаций (C). 

Важно осуществлять логирование, фиксируя для каждого клиента его ID, назначенную группу, факт выполнения целевых действий (например, первая операция), а также все входящие и исходящие коммуникации.

Параллельно ведётся операционный мониторинг для проверки равномерности распределения трафика между группами и отсутствия технических сбоев в сценариях взаимодействия.
Цель мониторинга - своевременно выявлять и устранять проблемы, угрожающие валидности эксперимента.

2.2. Контроль чистоты эксперимента
Это обеспечение условий, при которых различия в результатах между группами можно однозначно отнести к действию тестируемых факторов, а не к внешним воздействиям.

В нашем кейсе это означает изоляцию тестовых групп от непредусмотренных влияний. Необходимо гарантировать, что на клиентов в период теста не будут воздействовать другие маркетинговые кампании или изменения в стандартных процессах обслуживания, не связанные с экспериментом. 

Также важно контролировать кросс-контаминацию — например, чтобы клиент из группы с удалённым менеджером (A2) не мог по собственной инициативе обратиться к выделенному менеджеру (A1). Дополнительно отслеживаются ключевые внешние события (изменение законодательства, макроэкономические шоки), которые могут повлиять на отток всех групп одновременно.

2.3. Сбор и подготовка данных к анализу
Это этап накопления достаточного объёма данных согласно рассчитанной выборке и их трансформация в формат, пригодный для статистического анализа.

В нашем кейсе это означает ожидание достижения запланированного объёма выборки (≈53 тыс. клиентов) и истечения периода наблюдения в 4 месяца для каждого клиента для оценки первичной метрики (отток). На этом этапе данные из операционных систем (CRM, транзакционные базы, логи коммуникаций) агрегируются и преобразуются. Для каждого клиента формируется итоговая запись с признаком оттока на 4-м месяце и значениями вторичных метрик (количество транзакций, обороты, доход, количество негативных триггеров). Данные проверяются на целостность и очищаются от аномалий перед передачей в инструменты анализа.

#### 3. Этап анализа и интерпретации результатов

3.1. Сбор данных и проверка статистических предположений

На данном этапе проводится сбор данных по всем тестовым группам и проверка условий для применения статистических тестов (размер выборки, нормальность, гомогенность дисперсий).

Наш пример: После набора требуемого объема выборки и 4 месяцев наблюдения формируется таблица с долей оттока, доходом и другими метриками для каждой из 9 групп. Проверяется достаточность размера групп для применения параметрических тестов.

3.2. Статистический анализ

Оценивается значимость влияния факторов и их взаимодействий на целевую метрику.
Примеры методов статистического анализа:

- распространенным является метод ANOVA (дисперсионный анализ), который позволяет оценить главные эффекты каждого фактора (тип менеджера, активация, коммуникации). Однако у ANOVA есть ограничения: требует нормальности распределения и гомогенности дисперсий.
- при нормальном распределении, но гетероскедастичности дисперсий часто используют Welch-Anova.
- при нарушении условий нормальности на практике распространено использование непараметрических тестов (Крускала-Уоллиса).

Для нашего кейса может применяться дисперсионный анализ (ANOVA) на агрегированных данных (средние доли оттока по группам). Однако более корректным с методологической точки зрения является обобщённая линейная модель с биномиальным распределением (логистическая регрессия), которая работает с исходными бинарными данными на уровне клиента (отток/не отток) и корректно учитывает биномиальную природу метрики и зависимость дисперсии от величины пропорции.

Пример кода для логистической регрессии с оценкой p-value: (не запускаем, это просто пример кода, если бы у нас была таблица с данными df)

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# 1. Обучаем модель
model = sm.GLM.from_formula(
    'churn ~ C(factor_A) + C(factor_B) + C(factor_C)',
    data=df, #допустим, у нас есть таблица df с данными по признакам и оттоку 
    family=sm.families.Binomial()
).fit()

# 2.Таблица результатов с p-value
results_summary = model.summary2().tables[1]
print("Результаты модели:")
print(results_summary)

# 3. Проверяем значимость каждого фактора (тест Вальда)
print("\nПроверка статистической значимости")
alpha = 0.05

# Фактор A (сравниваем A2 и A3 с базой A1)
p_A2 = model.pvalues.get('C(factor_A)[T.A2]', 1.0)
p_A3 = model.pvalues.get('C(factor_A)[T.A3]', 1.0)
is_A_significant = p_A2 < alpha or p_A3 < alpha

print(f"Фактор A (Тип менеджера):")
print(f"  • p-value для A2 vs A1: {p_A2:.4f} {'✓' if p_A2 < alpha else '✗'}")
print(f"  • p-value для A3 vs A1: {p_A3:.4f} {'✓' if p_A3 < alpha else '✗'}")
print(f"  • Фактор значим? {'ДА' if is_A_significant else 'НЕТ'}")

# Фактор B
p_B2 = model.pvalues.get('C(factor_B)[T.B2]', 1.0)
p_B3 = model.pvalues.get('C(factor_B)[T.B3]', 1.0)
is_B_significant = p_B2 < alpha or p_B3 < alpha

print(f"\nФактор B (Мероприятие активации):")
print(f"  • p-value для B2 vs B1: {p_B2:.4f} {'✓' if p_B2 < alpha else '✗'}")
print(f"  • p-value для B3 vs B1: {p_B3:.4f} {'✓' if p_B3 < alpha else '✗'}")
print(f"  • Фактор значим? {'ДА' if is_B_significant else 'НЕТ'}")

# Фактор C
p_C2 = model.pvalues.get('C(factor_C)[T.C2]', 1.0)
p_C3 = model.pvalues.get('C(factor_C)[T.C3]', 1.0)
is_C_significant = p_C2 < alpha or p_C3 < alpha

print(f"\nФактор C (Частота коммуникаций):")
print(f"  • p-value для C2 vs C1: {p_C2:.4f} {'✓' if p_C2 < alpha else '✗'}")
print(f"  • p-value для C3 vs C1: {p_C3:.4f} {'✓' if p_C3 < alpha else '✗'}")
print(f"  • Фактор значим? {'ДА' if is_C_significant else 'НЕТ'}")

# 4. Определяем лучший уровень только для значимых факторов
best_levels = {}
base_churn = np.exp(model.params['Intercept']) / (1 + np.exp(model.params['Intercept']))

if is_A_significant:
    # Сравниваем коэффициенты
    coefs_A = {
        'A1': 0,  # базовая категория
        'A2': model.params.get('C(factor_A)[T.A2]', 0),
        'A3': model.params.get('C(factor_A)[T.A3]', 0)
    }
    # Находим уровень с минимальным коэффициентом (меньше вероятность оттока)
    best_A = min(coefs_A.items(), key=lambda x: x[1])[0]
    best_levels['A'] = best_A
else:
    best_levels['A'] = 'A1'  # оставляем текущий

if is_B_significant:
    coefs_B = {
        'B1': 0,
        'B2': model.params.get('C(factor_B)[T.B2]', 0),
        'B3': model.params.get('C(factor_B)[T.B3]', 0)
    }
    best_B = min(coefs_B.items(), key=lambda x: x[1])[0]
    best_levels['B'] = best_B
else:
    best_levels['B'] = 'B1'

if is_C_significant:
    coefs_C = {
        'C1': 0,
        'C2': model.params.get('C(factor_C)[T.C2]', 0),
        'C3': model.params.get('C(factor_C)[T.C3]', 0)
    }
    best_C = min(coefs_C.items(), key=lambda x: x[1])[0]
    best_levels['C'] = best_C
else:
    best_levels['C'] = 'C1'

# 5. Формируем и оцениваем комбинацию
recommended_combo = (best_levels['A'], best_levels['B'], best_levels['C'])

print(f"\nРекомнендации:")
print(f"Лучшие уровни значимых факторов: {best_levels}")
print(f"Рекомендуемая комбинация: {recommended_combo}")

# Предсказание для этой комбинации
pred_data = pd.DataFrame([recommended_combo], 
                         columns=['factor_A', 'factor_B', 'factor_C'])
predicted_prob = model.predict(pred_data)[0]

print(f"\nПрогноз оттока для комбинации: {predicted_prob:.2%}")
print(f"Снижение vs база ({base_churn:.2%}): {(base_churn - predicted_prob):.2%} п.п.")
print(f"Эффект значим? {'ДА' if (base_churn - predicted_prob) > 0.03 else 'НЕТ'} (MDE = 3%)")
```

3.3. Определение оптимальной комбинации и расчёт эффекта
На данном этапе осуществляется выбор комбинации уровней факторов с наилучшим значением метрики, расчёт величины эффекта и его доверительного интервала.

Наш пример: Для плана L9, который позволяет оценить только главные эффекты, определяется лучший уровень каждого значимого фактора на основе модели логистической регрессии. Формируется гипотеза для валидационного теста: комбинация лучших уровней (например, A1+B1+C1) может быть оптимальной. Эффект рассчитывается как предсказанное моделью снижение доли оттока, например, на 3.5 п.п. (95% ДИ: [1.2; 5.8]). Важно отметить, что эта комбинация не тестировалась непосредственно в рамках L9, поэтому требует отдельной проверки.

Можно такой вывод сделать для заказчика: "На основе значимых главных эффектов в плане L9 сформирована гипотеза о потенциально оптимальной комбинации (А1+В1+С1). Поскольку данная конкретная комбинация не тестировалась непосредственно в дробном плане (в L9 были другие сочетания этих факторов), а взаимодействия факторов не оценивались, требуется валидационный A/B тест для подтверждения её эффективности."

3.4. Принятие решения и генерация инсайтов
На данном этапе принимается решение о внедрении изменений на основе статистической значимости и бизнес-смысла, извлечение качественных инсайтов.

Наш пример: Если главные эффекты значимы (p < 0.05), ожидаемое снижение оттока превышает MDE (3 п.п.), а контрольные метрики не ухудшились — рекомендуется провести валидационный A/B тест гипотетически лучшей комбинации, поскольку план L9 позволяет оценить только главные эффекты, но не взаимодействия между факторами, которые могут повлиять на итоговый результат комбинации.

Инсайты могут включать качественные наблюдения (например, "бонус вызывает негативные отзывы в определённых группах"), но их статистическая значимость в рамках L9 не проверяема. Может потребоваться доп. исследование.

## **Вопрос №1**: 
В чем основное отличие MVT от A/B-тестирования?

**Ответ**:
A/B-тестирование:
 - может тестировать как одно изменение, так и несколько изменений одновременно (мультивариантный A/B-тест).
- при тестировании нескольких изменений одновременно ответ ограничен: "Версия B (со всеми изменениями) лучше/хуже/не имеет значимой разницы по сравнению с Версией A (без изменений)", но невозможно определить, какое именно изменение или комбинация изменений привела к результату.

MVT:
Предназначен для одновременного тестирования нескольких независимых факторов (элементов) с разными вариантами (уровнями) каждого.

Позволяет оценить:
- Главные эффекты — индивидуальное влияние каждого фактора
- при полном MVT (не дробном, как ранее приводили пример с L9)  — как факторы влияют друг на друга в комбинации 

Ответ теста может быть вида: "Фактор A даёт +X%, Фактор B даёт +Y%, но вместе они дают только +Z% (есть отрицательное взаимодействие)".

Таким образом, основное отличие MVT от A/B-тестирования — в глубине причинно-следственного анализа. A/B-тест, даже при тестировании пакета изменений, показывает лишь, какая версия лучше в целом (если есть разница). MVT специально спроектирован для оценки индивидуального влияния каждого фактора и их взаимодействий, что позволяет понять не только "что работает", но и "почему" и "как элементы влияют друг на друга".

## **Вопрос №2**: 
Какие метрики обычно используются для оценки эффективности MVT?

**Ответ**:
Для комплексной оценки эффективности MVT используется набор взаимодополняющих метрик, которые можно разделить на пять ключевых групп:

1. Основные бизнес-метрики
Это ключевые показатели, ради которых проводится тест. Рекомендовано выбирать 1 главную метрику, и остальные интересующие заказчика метрики выделить как вспомогательные. Примеры основных метрик:

- конверсия (например, конверсия просмотров в клики/ кликов в покупки)
- средний чек 
- отток клиентов

Принципы выбора: приоритетность, прямая связь с бизнес-целью, чувствительность, отсутствие конфликта с другими важными показателями.

2. Вспомогательные и поведенческие метрики: 
Помогают интерпретировать изменения основной метрики и контролировать побочные эффекты (например, VOC, проникновение продуктов, операционная нагрузка). Их цель — убедиться, что улучшение по основной метрике не достигнуто ценой ухудшения других бизнес-аспектов.

3. Статистические метрики и показатели достоверности

Критически важны для интерпретации результатов MVT.

Уровень значимости (p-value) и достоверность (обычно 95%)

Мощность теста: способность теста обнаружить реально существующий эффект. До запуска помогает рассчитать необходимую выборку.

Доверительный интервал: показывает диапазон, в котором с заданной вероятностью находится истинное значение улучшения (например, «увеличение конверсии на 5% ± 1,5%»).

Скорость сбора данных: Как быстро тест набирает статистическую мощность.

4. Метрики для анализа взаимодействий:
Это специфическая для MVT группа метрик. Они позволяют оценить, как разные изменения на странице влияют друг на друга.

- вклад фактора (главные эффекты): влияние каждого отдельного изменяемого элемента (например, только новая акция или только рост количества коммуникаций) на метрику в среднем по всем комбинациям.

- эффект взаимодействия: показывает, как комбинация двух или более изменений дает эффект, отличный от простой суммы их отдельных вкладов. Например, «новая акция улучшает результат только в совокупности с высоким уровнем кол-ва коммуникаций, а с низким уровнем не имеет влияния на результат».

5. Практические и бизнес-метрики эффективности теста
- размер обнаруженного эффекта (если он есть)
- скорость прохождения теста: как быстро был получен значимый результат.
- ROI тестирования

## **Вопрос №3**: 
Как определить количество групп в MVT?

**Ответ**:

Определение количества групп (вариантов, комбинаций) в MVT — это ключевой этап планирования, который напрямую влияет на сложность, длительность и результативность теста. 

Вот пошаговый алгоритм и важные принципы.

Основная формула:
Количество уникальных групп (вариантов) рассчитывается как произведение количества вариаций каждого изменяемого элемента.
N = v1 × v2 × v3 × ...
В нашем примере с оттоком клиентов ММБ мы тестировали 3 фактора по 3 вариации в каждом:
Общее количество групп: 3 × 3 × 3 = 27 уникальных комбинаций (включая контрольную, если она не выделена отдельно).

Приведем пошаговый алгоритм определения количества групп:

Шаг 1: Расчет полного факторного дизайна с использованием формулы выше. Это даст максимально возможное количество групп. В нашем примере — 27.
Это число показывает всю сложность предстоящего тест. 27 групп — это уже довольно сложный тест.

Шаг 2: Оценка ограничений и оптимизация.
Здесь при необходимости нужно применить фильтры, чтобы не тестировать всё подряд.
Чем больше групп, тем больше трафика и времени нужно для достижения статистической значимости для каждой.
Следует рассчитать размер выборки (используя MDE, базовое значение метрики, уровень значимости и мощности) и длительность теста. Затем, умножив это число на количество групп, получить общий требуемый объем трафика. В первой части сесионного задания приводили пример расчета.

Далее проводится оценка наличия ресурсов (время дизайнеров, разработчиков, аналитиков), возможности набрать рассчитанный объем выборки и бизнес-целесообразность ожидания результатов в рассчитанный срок.

Шаг 3: Принятие решения о дизайне теста (Full vs. Fractional)
Если полный факторный дизайн слишком велик, наиболее распространены 2 решения:

1. Упростить тест:

- уменьшить количество факторов. Возможно, стоит начать не с 3, а с 2 самых гипотетически сильных элементов.
- уменьшить количество вариаций. Например, вместо 3 вариаций протестировать 2

2. Использовать Дробный факторный дизайн.
Рассмотрено в первой части сесионного задания.

- плюс: существенно сокращает количество групп (как в нашем примере, с 27 до 9 групп) и необходимое время.
- минус: потеря информации: не получится измерить все эффекты взаимодействия (как элементы влияют друг на друга).
 Дробный факторный дизайн рекомендуется использовать, когда есть много факторов (5+), и полный тест невозможен, а заказчика интересует в первую очередь основное влияние каждого фактора.

## **Вопрос №4**: 
Какие риски связаны с проведением MVT?

**Ответ**:

Проведение MVT связано со значительными рисками, которые важно учитывать на этапе планирования. Их можно разделить на несколько категорий.

1. Статистические и методологические риски: 
- ложные открытия из-за множественного сравнения: Основной риск. При одновременном сравнения десятков вариантов с контролем резко растёт вероятность случайно найти «победителя» там, где его нет. Стандартные пороги значимости (95%) становятся ненадёжными, требуются поправки (Бонферрони, FDR).

- взаимодействия высокого порядка: в тесте с 4+ факторами могут возникать сложные взаимодействия. Их почти невозможно корректно проанализировать и интерпретировать на практике.

- риск нереалистичного планирования, ведущего к фактической недостаточной мощности. Из-за необходимости делить трафик между множеством групп, расчетный требуемый объем трафика часто оказывается недостижимым в разумные бизнес-сроки. Это вынуждает команды либо запускать заведомо маломощные тесты (высокий риск ложноотрицательных результатов), либо искусственно завышать MDE, упуская из виду потенциально важные, но умеренные эффекты.

- сложность интерпретации и внедрения эффектов взаимодействия. Победителем может оказаться специфическая комбинация элементов ([H2 + I3 + C1]), которая показывает результат за счёт их синергии, а не благодаря универсально сильным отдельным компонентам. Это усложняет генерализацию инсайтов и их масштабирование на другие части продукта

2. Ресурсные и операционные риски

- экспоненциальный рост сложности:к примеру,  4 элемента по 3 варианта = 81 группа. Проведение теста, анализ результатов требуют огромных трудозатрат. Растёт риск ошибок в реализации.

- длительное время теста: для набора статистической значимости по каждой из многих групп может потребоваться недели или месяцы. За это время рынок, сезонность или поведение пользователей могут измениться, делая результаты нерелевантными.

- высокий порог входа по трафику: MVT оправдан только для сайтов/приложений/продуктов/процессов... с очень высоким трафиком. Для малых и средних проектов часто нереализуем.

3. Риски интерпретации и принятия решений
- переобучение: победившая комбинация может быть оптимальна для этой конкретной аудитории и периода теста, но не хороша на будущее или другие сегменты.

- подмена стратегии тактикой: фокусировка на микрооптимизации процессов/продуктов с помощью MVT как на основном инструменте роста, что отвлекает ресурсы и внимание от проверки и исправления фундаментальных гипотез о продукте и рынке

- сложность внедрения инсайтов: даже успешный тест может дать противоречивые результаты (например, «красная кнопка работает лучше с коротким заголовком, но синяя — с длинным»). Решение о том, какой инсайт внедрять на других страницах, становится неочевидным.

- статистически победившая комбинация может противоречить бизнес-логике

4. Практические и технические риски
- сложность технической реализации десятков вариантов повышает вероятность багов, которые могут дискредитировать результаты теста для отдельных групп.

- ошибки в логике распределения: сложные системы распределения трафика между десятками групп могут давать сбои, приводя к несбалансированным выборкам и искажённым результатам.

- эффект новизны: краткосрочная положительная реакция на новые варианты, которая может исказить данные в начале теста и спровоцировать ложные выводы, особенно при мониторинге множества групп в MVT. Необходима выдержка и анализ данных за весь период теста

## **Вопрос №5**: 
Как интерпретировать взаимодействие факторов в MVT?

**Ответ**:

Интерпретация включает два ключевых шага:

1. Установить направление эффекта: как именно влияют факторы друг на друга — синергия (усиление) или антагонизм (ослабление)?

2. Сформулировать практический инсайт: какое конкретное бизнес-решение следует принять на основе этого взаимодействия? (Например: «Комбинировать только определённые варианты», «Сегментировать аудиторию», «Не использовать фактор X с фактором Y»).

Общий алгоритм:

1. Если взаимодействие статистически значимо (p-value < α):
 - посмотреть на график взаимодействия или таблицу средних значений для ключевых комбинаций.
 - задайть вопрос: «Изменяется ли влияние Фактора 1 в зависимости от уровня Фактора 2?»
 - сформулировать вывод: «Эффект от Фактора 1 является положительным/отрицательным при [значение А] Фактора 2, но становится незначимым/отрицательным/более сильным при [значение Б] Фактора 2.

2. Если взаимодействие незначимо: интерпретировать только главные эффекты каждого фактора по отдельности.

Пример интерпретации для нашего кейса с клиентами ММБ (допустим, провели полный MVT-тест, не дробный факторный анализ):
Допустим, мы обнаружили значимое взаимодействие B × C (Мероприятие по активации × Частота коммуникаций).

1. Данные:

B1 (бонус) + C3 (редкие коммуникации)	- отток 8%
B1 + C2 (умеренные коммуникации)	- отток 15%	
B1 + C1 (частые коммуникации)	- отток 21%	
B2 (розыгрыш призов) + C1	- отток 9%
B2 + C2	 - отток 20%	
B2 + C3	 - отток 25%	
B3 (бесплатные переводы) + C1 - отток 23%	
B3 + C2	 - отток 24%	
B3 + C3 - отток	23%	

2. Графический анализ (мысленный):

Для B1 (бонуса): линия будет нисходящей к C3 - чем реже коммуникации, тем лучше.
Для B2 (розыгрыш призов): линия будет нисходящей к C1 → чем чаще коммуникации, тем лучше.

Линии непараллельны - взаимодействие есть.

3. Словесная интерпретация и инсайт:

Эффект от типа мотивационного мероприятия (B) статистически значимо зависит от частоты коммуникаций (C).

Денежный бонус (B1) работает лучше всего при минимальном последующем контакте (C3). Вероятно, он воспринимается как простая и ясная ценность, а частые напоминания вызывают раздражение.

Розыгрыш призов (B2), наоборот, требует активной поддержки и частых коммуникаций (C1) для поддержания азарта и вовлечённости, без этого его эффективность падает.

Эффект от бесплатных переводов (B3) менее зависим от частоты контактов.

4. Практическое бизнес-решение:

Сегментировать стратегию удержания:

- для клиентов, получающих денежный бонус (B1), снизить частоту коммуникаций после его выплаты до минимальной (C3).

- для клиентов, участвующих в розыгрыше (B2), поддерживать высокую частоту контактов (C1) с напоминаниями, историями победителей и обратным отсчётом.

Это позволит снизить отток сильнее, чем применение любой единой частоты коммуникаций ко всем мероприятиям.

Итог: Интерпретация взаимодействия в MVT сводится к поиску условных правил вида «если применяем вариант X, то необходимо применять вариант Y». Это даёт не просто «победившую» комбинацию, а систему решений для разных сценариев.
