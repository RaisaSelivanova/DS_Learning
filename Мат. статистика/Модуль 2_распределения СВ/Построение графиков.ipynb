Задача 3.1. 
Постройте график плотности вероятности нормального распределения  на интервале [-5; 5]  с использованием модуля statsбиблиотеки SciPy.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Устанавливаем Seaborn стиль
sns.set_theme(style="whitegrid")

# Создаем данные
x = np.linspace(-5, 5, 1000)
y = stats.norm.pdf(x, 0, 1)  # Плотность N(0, 1)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='Плотность вероятности')

# Заливаем область под кривой с прозрачностью
plt.fill_between(x, y, alpha=0.5, color='skyblue')

# Вертикальные линии на ±σ (σ=1)
plt.axvline(x=1, color='red', linestyle='--', linewidth=1.5, label='±1σ')
plt.axvline(x=-1, color='red', linestyle='--', linewidth=1.5)

# Оформление
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.title('Плотность вероятности $\\mathcal{N}(0, 1)$', fontsize=16)
plt.legend()
plt.xlim([-5, 5])
plt.ylim([0, 0.45])

# Отображаем график
plt.tight_layout()
plt.show()
```

Задача 3.2.
Сгенерируйте выборку из 2000 значений из экспоненциального распределения с параметром лямбда = 1,5. Постройте гистограмму распределения

```python
# Устанавливаем Seaborn стиль
sns.set_theme(style="whitegrid")

# Создаем данные
x = np.random.exponential(scale = 1/1.5, size = 2000)

mean2 = 1/1.5 #мат ожидание

# Создаем график
plt.figure(figsize=(10, 6))
plt.hist(x, linewidth=2, label='Гистограмма распределения')


# Вертикальные линии на ±σ (σ=1)
plt.axvline(x=mean2, color='red', linestyle='--', linewidth=1.5, label='Мат ожидание')

# Оформление
plt.xlabel('Время до события (с)', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.title('Гисторграмма распределения СВ' , fontsize=16)
plt.legend()

# Отображаем график
plt.tight_layout()
plt.show()
```

```python
#Построим график теоретической плотности распределения

x_theoretical = np.linspace(0, 4, 2000)  # Равномерные точки от 0 до 4
y = stats.expon.pdf(x_theoretical, scale=1/1.5)  # Теоретическая плотность


# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x_theoretical, y, 'b-', linewidth=2, label='Плотность вероятности')

# Заливаем область под кривой с прозрачностью
plt.fill_between(x_theoretical, y, alpha=0.5, color='skyblue')

# Вертикальные линии на ±σ (σ=1)
plt.axvline(x=1/1.5, color='red', linestyle='--', linewidth=1.5, label='Мат ожидание')


# Оформление
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.title('Плотность вероятности $\\mathcal{N}(0, 1)$', fontsize=16)
plt.legend()

# Отображаем график
plt.tight_layout()
plt.show()
```

Задача 3.3. Биномиальное распределение

```python
# Параметры биномиального распределения
n = 15
p = 0.3

# Значения k от 0 до n
k_values = np.arange(0, n + 1)

# Рассчитываем вероятности P(k) для всех k от 0 до n
probabilities = stats.binom.pmf(k_values, n, p)

# Создаем график
plt.figure(figsize=(12, 6))

# Столбчатая диаграмма
bars = plt.bar(k_values, probabilities, alpha=0.7, 
               edgecolor='black', linewidth=1.2,
               color='skyblue', label=f'Bin(n={n}, p={p})')

# Маркеры точек поверх столбцов
plt.plot(k_values, probabilities, 'ro', markersize=8, 
         markerfacecolor='red', markeredgecolor='darkred',
         linestyle='None', label='Вероятности P(k)')

# Подписи и оформление
plt.xlabel('Число успехов (k)', fontsize=12)
plt.ylabel('Вероятность P(k)', fontsize=12)
plt.title(f'Биномиальное распределение: n={n}, p={p}', fontsize=14, pad=15)

plt.xticks(k_values)  # Все значения k на оси X
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.legend(loc='best')
plt.ylim(0, max(probabilities) * 1.2)  # Немного места сверху для подписей

plt.tight_layout()
plt.show()

# Основные характеристики
mean = stats.binom.mean(n, p)
var = stats.binom.var(n, p)
std = stats.binom.std(n, p)

print(f"\nМатематическое ожидание: E[X] = np = {mean:.2f}")
print(f"Дисперсия: Var[X] = np(1-p) = {var:.2f}")
print(f"Стандартное отклонение: σ = {std:.2f}")

# Наиболее вероятное значение (мода)
mode_index = np.argmax(probabilities)
print(f"Наиболее вероятное число успехов: k = {k_values[mode_index]}")
print(f"Вероятность моды: P({k_values[mode_index]}) = {probabilities[mode_index]:.4f}")
```

Задание 3.5.

```python
# Генерация данных
np.random.seed(42)
n = 100
x = np.random.uniform(0, 10, n)
epsilon = np.random.normal(0, 2, n)  # σ=2, так как σ²=4
y = 3 * x + epsilon

# DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Настройка стиля
sns.set_style("whitegrid")

# Создание графика
fig, ax = plt.subplots(figsize=(10, 6))

# Точечный график
sns.scatterplot(data=df, x='x', y='y', 
                alpha=0.7, ax=ax, s=50)

# Линия регрессии
sns.regplot(data=df, x='x', y='y', 
            scatter=False, ci=None,
            line_kws={'color': 'red', 'linewidth': 2.5},
            ax=ax)

# Оформление
ax.set_xlabel(r'$x$', fontsize=12)
ax.set_ylabel(r'$y = 3x + \mathcal{N}(0, 4)$', fontsize=12)
ax.set_title('Точечный график данных с линией регрессии', fontsize=14)

# Ограничение оси Y
ax.set_ylim(df['y'].min() - 2, df['y'].max() + 2)

# Добавим теоретическую линию
x_range = np.array([0, 10])
ax.plot(x_range, 3*x_range, 'g--', linewidth=2, 
        label=r'Истинная зависимость: $y = 3x$')

ax.legend()
plt.tight_layout()
plt.show()
```
