```python
import numpy as np
from scipy.stats import norm

# Шаг 1: Задаем известные параметры
sigma = 10.0      # Сигма известно!
n = 50            
x_bar = 125.8    
confidence = 0.95 

# Шаг 2: Вычисляем стандартное отклонение среднего
se = sigma / np.sqrt(n)

# Шаг 3: Находим z-критическое значение для 95% доверия
# Для двустороннего интервала: 1 - (1-0.95)/2 = 0.975
z_critical = norm.ppf(1 - (1 - confidence) / 2)

# Шаг 4: Вычисляем границы доверительного интервала
# Формула: x̄ ± z × (σ/√n)
lower_bound = x_bar - z_critical * se  # левая граница
upper_bound = x_bar + z_critical * se  # правая граница

# Шаг 5: Округляем левую границу до 2 знаков после запятой
lower_bound_rounded = round(lower_bound, 2)
upper_bound_rounded = round(upper_bound, 2)

print(f"Доверительный интервал: ({lower_bound:.3f}, {upper_bound:.3f})")
print(f"Левая граница (округленная до 2 знаков): {lower_bound_rounded}")
print(f"Правая граница (округленная до 2 знаков): {upper_bound_rounded}")
```

```python
import numpy as np
from scipy.stats import chi2

# Шаг 1: Данные из задачи
n = 17               # размер выборки
s2 = 16.0            # выборочная дисперсия (s² = 16)
confidence = 0.90    # уровень доверия 90%
alpha = 1 - confidence  # 0.10

# Шаг 2: Степени свободы
df = n - 1  # 16

# Шаг 3: Вычисляем (n-1)s²
n_minus_1_times_s2 = (n - 1) * s2  # 16 × 16 = 256

# Шаг 4: Критические значения χ²
# Для нижней границы: χ²(1-α/2, df) = χ²(0.95, 16)
# Для верхней границы: χ²(α/2, df) = χ²(0.05, 16)
chi2_lower = chi2.ppf(1 - alpha/2, df)  # χ²(0.95, 16)
chi2_upper = chi2.ppf(alpha/2, df)      # χ²(0.05, 16)

# Шаг 5: Доверительный интервал для дисперсии
lower_bound_var = n_minus_1_times_s2 / chi2_lower  # 256 / χ²(0.05, 16)
upper_bound_var = n_minus_1_times_s2 / chi2_upper  # 256 / χ²(0.95, 16)

# Шаг 6: Округление нижней границы до 1 знака после запятой
lower_bound_rounded = round(lower_bound_var, 1)

# Дополнительно: интервал для стандартного отклонения (√дисперсии)
lower_bound_std = np.sqrt(lower_bound_var)
upper_bound_std = np.sqrt(upper_bound_var)

print(f"Данные: n = {n}, s² = {s2}, доверительный уровень = {confidence*100}%")
print(f"Степени свободы: df = {df}")
print(f"(n-1)s² = {n_minus_1_times_s2}")
print(f"\nКритические значения χ²:")
print(f"χ²({alpha/2:.3f}, {df}) = {chi2_upper:.3f}")
print(f"χ²({1-alpha/2:.3f}, {df}) = {chi2_lower:.3f}")
print(f"\nДоверительный интервал для дисперсии σ²:")
print(f"({lower_bound_var:.3f}, {upper_bound_var:.3f})")
print(f"Нижняя граница (округлено до 1 знака): {lower_bound_rounded}")
print(f"\nДоверительный интервал для σ:")
print(f"({lower_bound_std:.3f}, {upper_bound_std:.3f})")
```

```python
import numpy as np
from scipy.stats import t

# Шаг 1: Данные
n1, x1, s1 = 25, 5.2, 1.1
n2, x2, s2 = 30, 4.6, 1.3
alpha = 0.05  # уровень значимости

# Шаг 2: Вычисление t-статистики
se = np.sqrt(s1**2/n1 + s2**2/n2)  # стандартная ошибка разности
t_stat = (x1 - x2) / se

# Шаг 3: Степени свободы (Welch-Satterthwaite)
num = (s1**2/n1 + s2**2/n2)**2  # числитель
den = (s1**2/n1)**2/(n1-1) + (s2**2/n2)**2/(n2-1)  # знаменатель
df = num / den

# Шаг 4: Двусторонний p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# Шаг 5: Критическое значение
t_critical = t.ppf(1 - alpha/2, df)

print("=== Welch t-test ===")
print(f"Группа 1: n={n1}, x̄={x1}, s={s1}")
print(f"Группа 2: n={n2}, x̄={x2}, s={s2}")
print(f"\nРазность средних: {x1} - {x2} = {x1-x2:.2f}")
print(f"Стандартная ошибка: √({s1:.2f}²/{n1} + {s2:.2f}²/{n2}) = {se:.4f}")
print(f"t-статистика: {t_stat:.4f}")
print(f"\nСтепени свободы (Welch):")
print(f"  Числитель: ({s1:.2f}²/{n1} + {s2:.2f}²/{n2})² = {num:.6f}")
print(f"  Знаменатель: ({s1:.2f}²/{n1})²/({n1-1}) + ({s2:.2f}²/{n2})²/({n2-1}) = {den:.6f}")
print(f"  df = {df:.2f}")
print(f"\nКритическое значение t({1-alpha/2:.3f}, {df:.2f}) = {t_critical:.4f}")
print(f"Двусторонний p-value = {p_value:.6f}")
print(f"Уровень значимости α = {alpha}")

# Шаг 6: Проверка гипотезы
if abs(t_stat) > t_critical:
    print("\n✅ ОТВЕРГАЕМ H₀: |t-наблюдаемое| > t-критическое")
    print("✅ Принимаем H₁: μ₁ ≠ μ₂ (средние статистически различны)")
else:
    print("\n✅ ПРИНИМАЕМ H₀: |t-наблюдаемое| ≤ t-критическое")
    print("✅ Принимаем H₀: μ₁ = μ₂ (нет статистически значимых различий)")
```

```python
import numpy as np
from scipy.stats import f

# Шаг 1: Данные
n_x, s2_x = 20, 2.25  # выборка X
n_y, s2_y = 18, 1.44  # выборка Y
alpha = 0.05

# Шаг 2: Определяем большую и меньшую дисперсии
if s2_x >= s2_y:
    s2_larger, s2_smaller = s2_x, s2_y
    n_larger, n_smaller = n_x, n_y
else:
    s2_larger, s2_smaller = s2_y, s2_x
    n_larger, n_smaller = n_y, n_x

# Шаг 3: F-статистика (большая дисперсия / меньшая)
F_stat = s2_larger / s2_smaller

# Шаг 4: Степени свободы
df1 = n_larger - 1  # для большей дисперсии
df2 = n_smaller - 1  # для меньшей дисперсии

# Шаг 5: Двусторонний p-value
# Для двустороннего теста умножаем на 2 меньший из двух хвостов
p_value = 2 * min(f.cdf(F_stat, df1, df2), 1 - f.cdf(F_stat, df1, df2))

# Шаг 6: Критические значения
F_crit_lower = f.ppf(alpha/2, df1, df2)
F_crit_upper = f.ppf(1 - alpha/2, df1, df2)

print("=== F-test для равенства дисперсий ===")
print(f"Выборка X: n={n_x}, s²={s2_x}")
print(f"Выборка Y: n={n_y}, s²={s2_y}")
print(f"\nБольшая дисперсия: {s2_larger} (n={n_larger})")
print(f"Меньшая дисперсия: {s2_smaller} (n={n_smaller})")
print(f"F = s²_большая / s²_меньшая = {F_stat:.4f}")
print(f"\nСтепени свободы: df1={df1}, df2={df2}")
print(f"Критические значения F({alpha/2}, {df1}, {df2}) = {F_crit_lower:.4f}")
print(f"                    F({1-alpha/2}, {df1}, {df2}) = {F_crit_upper:.4f}")
print(f"Двусторонний p-value = {p_value:.6f}")
print(f"Уровень значимости α = {alpha}")

# Шаг 7: Проверка гипотезы
if F_stat < F_crit_lower or F_stat > F_crit_upper:
    print(f"\n✅ ОТВЕРГАЕМ H₀: F-наблюдаемое вне интервала ({F_crit_lower:.4f}, {F_crit_upper:.4f})")
    print("✅ Принимаем H₁: σₓ² ≠ σᵧ² (дисперсии статистически различны)")
else:
    print(f"\n✅ ПРИНИМАЕМ H₀: F-наблюдаемое внутри интервала ({F_crit_lower:.4f}, {F_crit_upper:.4f})")
    print("✅ Принимаем H₀: σₓ² = σᵧ² (нет статистически значимых различий в дисперсиях)")
```

```python
# Доверительный интервал для отношения дисперсий
ci_lower = F_stat / F_crit_upper
ci_upper = F_stat * F_crit_upper

print(f"\n95% доверительный интервал для отношения дисперсий σₓ²/σᵧ²:")
print(f"({ci_lower:.4f}, {ci_upper:.4f})")
print(f"Гипотеза σₓ²/σᵧ² = 1 попадает в интервал? {ci_lower <= 1 <= ci_upper}")
```

```python
import numpy as np
from scipy.stats import chi2

# Шаг 1: Таблица наблюдаемых частот
observed = np.array([[30, 20],
                     [10, 40]])

# Шаг 2: Суммы
row_sums = observed.sum(axis=1)  # [50, 50]
col_sums = observed.sum(axis=0)  # [40, 60]
total = observed.sum()  # 100

print("Наблюдаемые частоты (O):")
print(observed)
print(f"\nСуммы строк: {row_sums}")
print(f"Суммы столбцов: {col_sums}")
print(f"Общая сумма: {total}")

# Шаг 3: Ожидаемые частоты при независимости
expected = np.outer(row_sums, col_sums) / total

print("\nОжидаемые частоты (E) при независимости:")
print(expected)

# Шаг 4: Вычисление χ² статистики
chi2_stat = np.sum((observed - expected)**2 / expected)

print(f"\nχ² = Σ[(Oᵢⱼ - Eᵢⱼ)² / Eᵢⱼ]")
print("По ячейкам:")
for i in range(2):
    for j in range(2):
        o = observed[i, j]
        e = expected[i, j]
        contrib = (o - e)**2 / e
        print(f"  Ячейка ({i},{j}): ({o} - {e:.1f})² / {e:.1f} = {contrib:.3f}")

print(f"\nСумма (χ² статистика) = {chi2_stat:.4f}")

# Шаг 5: Степени свободы
df = (2 - 1) * (2 - 1)  # 1

# Шаг 6: p-value
p_value = 1 - chi2.cdf(chi2_stat, df)

# Шаг 7: Критическое значение
alpha = 0.05
chi2_critical = chi2.ppf(1 - alpha, df)

print(f"\nСтепени свободы: df = {df}")
print(f"Критическое значение χ²(0.95, {df}) = {chi2_critical:.4f}")
print(f"p-value = {p_value:.6f}")
print(f"Уровень значимости α = {alpha}")

# Шаг 8: Проверка гипотезы
if chi2_stat > chi2_critical:
    print(f"\n✅ ОТВЕРГАЕМ H₀: χ² наблюдаемое ({chi2_stat:.4f}) > χ² критического ({chi2_critical:.4f})")
    print("✅ Принимаем H₁: есть статистически значимая связь между строками и столбцами")
else:
    print(f"\n✅ ПРИНИМАЕМ H₀: χ² наблюдаемое ({chi2_stat:.4f}) ≤ χ² критического ({chi2_critical:.4f})")
    print("✅ Принимаем H₀: нет статистически значимой связи (независимость)")
```

```python
import numpy as np
from scipy.stats import t

# Шаг 1: Данные
n = 40
r = 0.35
alpha = 0.05

# Шаг 2: Вычисление t-статистики
t_stat = r * np.sqrt(n - 2) / np.sqrt(1 - r**2)

# Шаг 3: Степени свободы
df = n - 2  # 38

# Шаг 4: Двусторонний p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# Шаг 5: Критическое значение
t_critical = t.ppf(1 - alpha/2, df)

# Шаг 6: Округление t-статистики до 2 знаков после запятой
t_stat_rounded = round(t_stat, 2)

print("=== t-тест для коэффициента корреляции ===")
print(f"Размер выборки: n = {n}")
print(f"Выборочная корреляция: r = {r}")
print(f"\nt-статистика: r√(n-2)/√(1-r²) = {r}√({n-2})/√(1-{r}²)")
print(f"t = {t_stat:.4f}")
print(f"t (округлено до 2 знаков): {t_stat_rounded}")
print(f"\nСтепени свободы: df = n - 2 = {df}")
print(f"Критическое значение t({1-alpha/2:.3f}, {df}) = ±{t_critical:.4f}")
print(f"Двусторонний p-value = {p_value:.6f}")
print(f"Уровень значимости α = {alpha}")

# Шаг 7: Проверка гипотезы
if abs(t_stat) > t_critical:
    print(f"\n✅ ОТВЕРГАЕМ H₀: |t-наблюдаемое| ({abs(t_stat):.4f}) > t-критическое ({t_critical:.4f})")
    print("✅ Принимаем H₁: ρ ≠ 0 (корреляция статистически значима)")
else:
    print(f"\n✅ ПРИНИМАЕМ H₀: |t-наблюдаемое| ({abs(t_stat):.4f}) ≤ t-критическое ({t_critical:.4f})")
    print("✅ Принимаем H₀: ρ = 0 (нет статистически значимой корреляции)")
```

```python
import numpy as np
from scipy.stats import f

# Шаг 1: Данные
k = 3           # количество групп
n_per_group = 10 # наблюдений в каждой группе
SSB = 180       # межгрупповая сумма квадратов
SSW = 90        # внутригрупповая сумма квадратов

# Шаг 2: Общее количество наблюдений
N = k * n_per_group  # 30

# Шаг 3: Степени свободы
df_between = k - 1      # 2
df_within = N - k       # 27
df_total = N - 1        # 29

# Шаг 4: Средние квадраты
MSB = SSB / df_between  # 90
MSW = SSW / df_within   # 90/27 ≈ 3.3333

# Шаг 5: F-статистика
F_stat = MSB / MSW

# Шаг 6: Проверка вычислений
print("=== Односторонний ANOVA (однофакторный) ===")
print(f"Количество групп: k = {k}")
print(f"Наблюдений в группе: n = {n_per_group}")
print(f"Всего наблюдений: N = k × n = {N}")
print(f"\nСуммы квадратов:")
print(f"Межгрупповая (SSB) = {SSB}")
print(f"Внутригрупповая (SSW) = {SSW}")
print(f"Общая (SST) = SSB + SSW = {SSB + SSW}")
print(f"\nСтепени свободы:")
print(f"Межгрупповая: df_between = k-1 = {df_between}")
print(f"Внутригрупповая: df_within = N-k = {df_within}")
print(f"Общая: df_total = N-1 = {df_total}")
print(f"\nСредние квадраты:")
print(f"MSB = SSB/df_between = {SSB}/{df_between} = {MSB}")
print(f"MSW = SSW/df_within = {SSW}/{df_within} = {MSW:.4f}")
print(f"\nF-статистика = MSB/MSW = {MSB}/{MSW:.4f} = {F_stat:.4f}")
print(f"F-статистика (округлено до целого) = {int(np.round(F_stat))}")

# Шаг 7: Проверка гипотезы (для полноты)
alpha = 0.05
F_critical = f.ppf(1 - alpha, df_between, df_within)
p_value = 1 - f.cdf(F_stat, df_between, df_within)

print(f"\nПроверка гипотезы (α = {alpha}):")
print(f"Критическое значение F({1-alpha}, {df_between}, {df_within}) = {F_critical:.4f}")
print(f"p-value = {p_value:.10f}")
print(f"Отвергаем H₀ (все средние равны)? {F_stat > F_critical}")
```
