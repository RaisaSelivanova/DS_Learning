```python
import sympy as sp

x = sp.symbols('x', positive=True)
C = sp.symbols('C')

# Вычисляем интеграл
integral = sp.integrate(sp.exp(-4*x+2), (x, 0, sp.oo))
print(f"∫exp(-4x+2)dx от 0 до ∞ = {integral}")

# Решаем уравнение C * integral = 1
solution = sp.solve(C * integral - 1, C)
C_value = solution[0]
print(f"Символьное значение C = {C_value}")

# Численное значение с округлением до 2 знаков
C_numeric = float(C_value)
print(f"Численное значение C = {C_numeric:.2f}")
```

```python
import scipy.stats as stats

# Квантили
z1 = stats.norm.ppf(0.2)  # ≈ -0.8416
z2 = stats.norm.ppf(0.6)  # ≈ 0.2533

# Система уравнений:
# (10 - mu)/sigma = z1
# (15 - mu)/sigma = z2

sigma = (15 - 10) / (z2 - z1)
mu = 10 - z1 * sigma

print(f"mu = {mu:.3f}")
print(f"sigma = {sigma:.3f}")
print(f"sigma (округление до 2 знаков) = {sigma:.2f}")
```

```python
import math
from scipy import stats

n = 10
s = 0.8
nu = n - 1
alpha = 0.05
chi2_low = stats.chi2.ppf(alpha/2, nu)       # χ²(0.025,9)
chi2_high = stats.chi2.ppf(1 - alpha/2, nu)  # χ²(0.975,9)
var_L = (nu * s**2) / chi2_high
var_R = (nu * s**2) / chi2_low
sigma_L = math.sqrt(var_L)
sigma_R = math.sqrt(var_R)
print(f"Нижняя граница σ: {sigma_L:.4f}")
print(f"Верхняя граница σ: {sigma_R:.4f}")
print(f"Округление до 2 знаков: {sigma_L:.2f};{sigma_R:.2f}")
```

```python
from scipy import stats
import numpy as np

before = [2, 2, 5, 7, 8, 9, 10, 10, 10, 11]
after  = [5, 4, 6, 7, 11, 9, 9, 13, 11, 10]
d = np.array(after) - np.array(before)

print("Разности:", d)
print("Среднее разностей:", np.mean(d))
print("Медиана разностей:", np.median(d))

# Тест Шапиро-Уилка
shapiro_stat, shapiro_p = stats.shapiro(d)
print(f"\nШапиро-Уилк: статистика = {shapiro_stat:.4f}, p-value = {shapiro_p:.4f}")
```

```python
import numpy as np
from scipy import stats

class1 = np.array([79, 88, 39, 40, 40, 90, 39, 79, 75, 53])
class2 = np.array([62, 64, 87, 49, 72, 65, 63, 56, 77, 41])
shapiro1 = stats.shapiro(class1)
shapiro2 = stats.shapiro(class2)
print(f"Шапиро-Уилк, класс 1: W={shapiro1.statistic:.4f}, p={shapiro1.pvalue:.4f}")
print(f"Шапиро-Уилк, класс 2: W={shapiro2.statistic:.4f}, p={shapiro2.pvalue:.4f}")
var1 = np.var(class1, ddof=1)
var2 = np.var(class2, ddof=1)
print(f"Дисперсия 1: {var1:.4f}")
print(f"Дисперсия 2: {var2:.4f}")
F_obs = var1 / var2  # var1 > var2
df1 = len(class1) - 1
df2 = len(class2) - 1

# Двусторонний p-value
p_one_sided = stats.f.sf(F_obs, df1, df2)  # вероятность получить значение больше F_obs
p_two_sided = 2 * p_one_sided
print(f"F = {F_obs:.5f}")
print(f"p-value (двусторонний) = {p_two_sided:.5f}")
print(f"p-value до 4 знаков: {p_two_sided:.4f}")
```

```python
bartlett_stat, bartlett_p = stats.bartlett(class1, class2)
levene_stat, levene_p = stats.levene(class1, class2)
print(f"\nБартлетт: статистика = {bartlett_stat:.5f}, p = {bartlett_p:.5f}")
print(f"Левен: статистика = {levene_stat:.5f}, p = {levene_p:.5f}")
```

```python
import numpy as np
from scipy import stats

regular = np.array([4.17, 5.58, 5.18, 6.11, 4.50, 4.61, 5.17, 4.53, 5.33, 5.14])
partial = np.array([4.52, 4.18, 4.84, 5.89, 3.57, 3.91, 5.96, 6.09, 4.74, 5.26])
none    = np.array([4.81, 4.17, 4.41, 3.59, 5.87, 3.83, 6.03, 4.89, 4.32, 4.69])
shapiro_reg = stats.shapiro(regular)
shapiro_par = stats.shapiro(partial)
shapiro_non = stats.shapiro(none)

print("Шапиро-Уилк:")
print(f"Регулярный: W={shapiro_reg.statistic:.4f}, p={shapiro_reg.pvalue:.4f}")
print(f"Частичный:  W={shapiro_par.statistic:.4f}, p={shapiro_par.pvalue:.4f}")
print(f"Без ухода:  W={shapiro_non.statistic:.4f}, p={shapiro_non.pvalue:.4f}")
lev_stat, lev_p = stats.levene(regular, partial, none)
print(f"\nКритерий Левена: F={lev_stat:.4f}, p={lev_p:.4f}")

F_stat, p_val = stats.f_oneway(regular, partial, none)
print(f"\nОднофакторный ANOVA:")
print(f"F = {F_stat:.6f}")
print(f"p-value = {p_val:.6f}")
print(f"p-value до 4 знаков: {p_val:.4f}")
```

```python
import numpy as np

area = np.array([2310, 2333, 2356, 2379, 2402, 2425, 2448, 2471, 2494, 2517,
                 2540, 2543, 2602, 2615, 2625])
price = np.array([142000, 144000, 151000, 151000, 139000, 169000, 126000, 142000,
                  163000, 169000, 149000, 152000, 156000, 162000, 172000])

r = np.corrcoef(area, price)[0, 1]
print(f"Коэффициент корреляции Пирсона: {r:.6f}")
print(f"Округление до 4 знаков: {r:.4f}")
```

```python
mean_x = np.mean(area)
mean_y = np.mean(price)

cov = np.sum((area - mean_x) * (price - mean_y))
std_x = np.sqrt(np.sum((area - mean_x)**2))
std_y = np.sqrt(np.sum((price - mean_y)**2))

r_manual = cov / (std_x * std_y)
print(f"Проверка вручную: r = {r_manual:.6f}")
```

```python
import statsmodels.api as sm
offices = np.array([2, 2, 3, 3, 2, 4, 2, 2, 3, 4, 2, 3, 4, 2, 4])
entrances = np.array([2, 2, 2, 2, 3, 3, 2, 2, 3, 4, 3, 4, 2, 3, 4])
age = np.array([20, 12, 33, 43, 53, 23, 37, 34, 23, 55, 22, 41, 29, 17, 20])
def standardize(arr):
    return (arr - np.mean(arr)) / np.std(arr, ddof=1)

area_z = standardize(area)
offices_z = standardize(offices)
entrances_z = standardize(entrances)
age_z = standardize(age)
price_z = standardize(price)

# Матрица предикторов
X = np.column_stack((area_z, offices_z, entrances_z, age_z))
X = sm.add_constant(X)  # добавляем константу (но для стандартизованных данных она будет ≈0)

# Модель
model = sm.OLS(price_z, X).fit()
print(model.summary())

# Коэффициенты без константы
coeffs = model.params[1:]  # убираем константу
print("\nСтандартизованные коэффициенты (бета):")
for i, coef in enumerate(coeffs):
    print(f"β{i+1} = {coef:.6f}")

max_coef = np.max(np.abs(coeffs))
print(f"\nМаксимальный по модулю коэффициент: {max_coef:.6f}")
print(f"Округление до 4 знаков: {max_coef:.4f}")
```

```python
import numpy as np
import statsmodels.api as sm

area = np.array([2310, 2333, 2356, 2379, 2402, 2425, 2448, 2471, 2494, 2517,
                 2540, 2543, 2602, 2615, 2625])
offices = np.array([2, 2, 3, 3, 2, 4, 2, 2, 3, 4, 2, 3, 4, 2, 4])
entrances = np.array([2, 2, 2, 2, 3, 3, 2, 2, 3, 4, 3, 4, 2, 3, 4])
age = np.array([20, 12, 33, 43, 53, 23, 37, 34, 23, 55, 22, 41, 29, 17, 20])
price = np.array([142000, 144000, 151000, 151000, 139000, 169000, 126000, 142000,
                  163000, 169000, 149000, 152000, 156000, 162000, 172000])

X = np.column_stack((area, offices, entrances, age))
X = sm.add_constant(X)
model = sm.OLS(price, X).fit()

print(f"R² = {model.rsquared:.4f}")
print(f"F-статистика = {model.fvalue:.4f}")
print(f"p-value F-теста = {model.f_pvalue:.6f}")

from scipy import stats
alpha = 0.05
df1 = 4
df2 = 10
F_tab = stats.f.ppf(1 - alpha, df1, df2)
print(f"Fтабл (α=0.05, df1=4, df2=10) = {F_tab:.4f}")
model.fvalue - F_tab
```
